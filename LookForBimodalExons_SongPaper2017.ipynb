```python
import pandas as pd
```

```python
modalities = pd.read_csv("../data/GSE85908_modalities_tidy.csv")
modalities.head()
```

```python
# Remove first column
modalities = modalities.iloc[:,[1,2,3]]
modalities.head()
```

```python
# Get all bimodal events
bimodal_events = modalities[modalities["modality"]=="bimodal"]
print bimodal_events.shape
bimodal_events.head()
```

```python
# Get MN bimodal events
MN_bimodal_events = bimodal_events[bimodal_events["phenotype"]=="MN"]
print MN_bimodal_events.shape
```

```python
# Get NPC bimodal events
NPC_bimodal_events = bimodal_events[bimodal_events["phenotype"]=="NPC"]
print NPC_bimodal_events.shape
```

```python
# Get iPSC bimodal events
iPSC_bimodal_events = bimodal_events[bimodal_events["phenotype"]=="iPSC"]
print iPSC_bimodal_events.shape
```

```python
iPSC_bimodal_events.head()
```

```python
# Get the alternatively included exon coordinates for iPSC bimodal events
iPSC_altexons = iPSC_bimodal_events["event_id"].str.split("@").str[1].str.split("exon:").str[1]
#print iPSC_altexons.head()
iPSC_altexons_chrom = iPSC_altexons.str.split(":").str[0]
iPSC_bimodal_events = iPSC_bimodal_events.assign(alt_exon_chrom=iPSC_altexons_chrom)
#print iPSC_altexons_chrom.head()
iPSC_altexons_start = iPSC_altexons.str.split(":").str[1].str.split("-").str[0]
iPSC_bimodal_events = iPSC_bimodal_events.assign(alt_exon_start=iPSC_altexons_start)
#print iPSC_altexons_start.head()
iPSC_altexons_end = iPSC_altexons.str.split(":").str[1].str.split("-").str[1]
iPSC_bimodal_events = iPSC_bimodal_events.assign(alt_exon_end=iPSC_altexons_end)
#print iPSC_altexons_end.head()
iPSC_altexons_strand = iPSC_altexons.str.split(":").str[2]
iPSC_bimodal_events = iPSC_bimodal_events.assign(alt_exon_strand=iPSC_altexons_strand)
#print iPSC_altexons_strand.head()
iPSC_bimodal_events.head()
```

```python
iPSC_bimodal_events_sorted = iPSC_bimodal_events.sort_values(by=["alt_exon_chrom","alt_exon_start","alt_exon_end"])
iPSC_bimodal_events_sorted.head()
```

```python
# Write bed file 
iPSC_bimodal_events_sorted.iloc[:,[3,4,5,6]].to_csv("../processed_files/GSE85908_modalities_iPSCs_bimodalExons.bed",sep="\t",header=False,index=False)
```
